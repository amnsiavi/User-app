```python
def data_pipeline(data,jwt_token,url):
    
    headers = {

    'Authorization': 'Bearer {}'.format(jwt_token),
    'Content-Type': 'application/json'
    }
   
    for data in data:
        count = 1
        print('Storing data {}'.format(count))
        time.sleep(2)
        response = requests.post(url,json={
            'name':data['name'],'username':data['username'],
            'email':data['email'], 'address':{
              'street':data['address']['street'],
                'suite':data['address']['suite'],
                'city':data['address']['city'],
                'zip_code':data['address']['zipcode'],
                'geo_location':{'lat':data['address']['geo']['lat'],'lng':data['address']['geo']['lng']},
            },
            'phone':data['phone'], 'website':f"https://www.{data['website']}.com",'company':{
              'name':data['company']['name'], 'catchPhrase':data['company']['catchPhrase'],'bs':data['company']['bs']  
            },
        },headers=headers)
        if response.status_code == 200:
            print(f'data : {count} completed')
            count = count+1
        else:
            print('Error:', response.status_code)
            break
        
```
